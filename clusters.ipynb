In this project, you'll investigate properties of mushrooms. This classic dataset contains over 8000 observations, where each mushroom is described by a variety of features like color, odor, etc., and the target variable is an indicator for whether the mushroom is poisonous. Since all the observations are categorical, I've binarized the feature space. Look at the feature_names below to see all 126 binary names.

You'll start by running PCA to reduce the dimensionality from 126 down to 2 so that you can easily visualize the data. In general, PCA is very useful for visualization (though sklearn.manifold.tsne is known to produce better visualizations). Recall that PCA is a linear transformation. The 1st projected dimension is the linear combination of all 126 original features that captures as much of the variance in the data as possible. The 2nd projected dimension is the linear combination of all 126 original features that captures as much of the remaining variance as possible. The idea of dense low dimensional representations is crucial to machine learning!

Once you've projected the data to 2 dimensions, you'll experiment with clustering using KMeans and density estimation with Gaussian Mixture Models.
```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.mixture import GMM
from matplotlib.colors import LogNorm
```

```python
### Common functions ###

def print_div_line(text, line_char):
    print text + line_char*(80-len(text))
```

Load feature names.

```python
feature_names = []
lines = open('mushroom.map').read().splitlines()
for line in lines:
    [index, name, junk] = line.split()
    feature_names.append(name)
print 'Loaded feature names:', len(feature_names)
```

Load data. The data is sparse in the input file, but there aren't too many features, so we'll use a dense representation, which is supported by all sklearn objects.

```python
X, Y = [], []
lines = open('mushroom.data').read().splitlines()
for line in lines:
    items = line.split()
    Y.append(int(items.pop(0)))
    x = np.zeros(len(feature_names))
    for item in items:
        feature = int(item.split(':')[0])
        x[feature] = 1
    X.append(x)

# Convert these lists to numpy arrays.
X = np.array(X)
Y = np.array(Y)

# Split into train and test data.
train_data, train_labels = X[:7000], Y[:7000]
test_data, test_labels = X[7000:], Y[7000:]

# Check that the shapes look right.
print train_data.shape, test_data.shape
```

```python
def P1():
### STUDENT START ###

    # Run a PCA analysis with 50 components
    pca50 = PCA(n_components=50)
    pca50.fit(train_data)
    sorted_variances = np.sort(pca50.explained_variance_ratio_)[::-1]
    
    # Generate a table for the variance and cumulative variance for each compoent
    # sorted from highest variance to lowest
    print 'Principal Component\tVariance\tCumulative'
    print_div_line('', '_')
    cumulative_variance = 0
    for component in range(sorted_variances.shape[0]):
        cumulative_variance += sorted_variances[component]
        print component+1, '\t\t\t', \
                round(sorted_variances[component]*100, 2), '%\t\t', \
                round(cumulative_variance*100, 2), '%'

### STUDENT END ###

P1()
```

(5) Fit two GMMs, one for the positive examples and one for the negative examples in your 2d projected data. Predict the test examples by choosing the label for which the model gives a larger probability (use GMM.score). What GMM settings give the best prediction accuracy? How many PCA components do you need to give 100% accuracy?

```python
def P5():
### STUDENT START ###

    def gmm_predict_test(pos_train_data, neg_train_data, test_data, 
                         n_mix_components, cov_matrix_type):
        """
        Use 2 GMMs, one for positive training data, one for negative training data,
        to predict the labels for the test data, and return the accuracy of the predictions
        """
        
        # GMM for positive training data cases
        pos_clf = GMM(n_mix_components, cov_matrix_type)
        pos_clf.fit(pos_train_data)
        
        # GMM for negative training data cases
        neg_clf = GMM(n_mix_components, cov_matrix_type)
        neg_clf.fit(neg_train_data)
        
        # Calculate scores for test data using both models
        pos_scores = pos_clf.score(test_data)
        neg_scores = neg_clf.score(test_data)
        
        # Generate a predicted label array determined from the model that produced the highest
        # score for each test data case
        condlist = [pos_scores >= neg_scores, neg_scores > pos_scores]
        choicelist = [1, 0]
        pred_labels = np.select(condlist, choicelist)
        
        # Calculate accuracy of predicted label array against actual labels
        accuracy = (pred_labels[pred_labels == test_labels].shape[0] + 0.0) / test_labels.shape[0]
        return accuracy
        
        
    # Use PCA analysis to generate 2d train and test data
    pca2 = PCA(n_components=2)
    train_data2 = pca2.fit_transform(train_data)
    positive_cases = train_data2[train_labels == 1, :]
    negative_cases = train_data2[train_labels == 0, :]
    test_data2 = pca2.transform(test_data)
    
    # Determine best GMM settings for number of components and covariance matrix type
    max_mix_n = 4
    cov_types = ['spherical', 'diag', 'tied', 'full']
    accuracy_matrix = np.zeros((max_mix_n*len(cov_types), 3))    
    for n in range(1, max_mix_n+1):
        for cov_i in range(len(cov_types)):
            accuracy = gmm_predict_test(positive_cases, negative_cases, test_data2, 
                                        n, cov_types[cov_i])
            accuracy_matrix[max_mix_n*(n-1)+cov_i,0] = n
            accuracy_matrix[max_mix_n*(n-1)+cov_i,1] = cov_i
            accuracy_matrix[max_mix_n*(n-1)+cov_i,2] = accuracy
    
    best_settings = accuracy_matrix[accuracy_matrix[:,2] == max(accuracy_matrix[:,2])].flatten()
    print_div_line('Best GMM settings', '_')
    print '\tNumber of components: ', int(best_settings[0])
    print '\tCovariance matrix type: ', cov_types[int(best_settings[1])]
    print '\tAccuracy: ', round(best_settings[2]*100, 2), '%'
    print 
    
    # Determine PCA components for 100% accuracy
    max_pca_n = 126
    best_pca_n = 0
    for n in range(3, max_pca_n+1):
        # Use PCA analysis to generate test and train data sets with n features
        pca_n = PCA(n_components=n)
        train_data_n = pca_n.fit_transform(train_data)
        positive_cases = train_data_n[train_labels == 1, :]
        negative_cases = train_data_n[train_labels == 0, :]
        test_data_n = pca_n.transform(test_data)
        
        # Calculate accuracy for PCA with n-components using best settings for GMM
        accuracy = gmm_predict_test(positive_cases, negative_cases, test_data_n, 
                                    int(best_settings[0]), cov_types[int(best_settings[1])])
        #print 'pca_n: ', n, '\taccuracy: ', accuracy
        
        # Stop when accuracy has reached 100%
        if (round(accuracy*100, 2) >= 100.0):
            best_pca_n = n
            break;
    
    print_div_line('Minimum PCA components for 100% accuracy', '_')
    print '\t', best_pca_n

### STUDENT END ###

P5()
```
